# Constrained LQR CUDA validation and benchmark

Start a fresh Kaggle notebook with a **Tesla P100 GPU** and Internet access. This notebook records the host, compiler, and GPU specifications; builds the pure-FP64 CPU and CUDA implementations; runs CPU, emulation, native-CUDA, and Compute Sanitizer tests; and benchmarks every horizon through $N=16384$. The benchmark neither installs nor times JAX.

```python
%%bash
set -euo pipefail

repo=/kaggle/working/constrained_lqr_elimination
revision=optional-cuda-backend

if [[ ! -d "${repo}/.git" ]]; then
  git clone --filter=blob:none --no-checkout \
    https://github.com/joaospinto/constrained_lqr_elimination.git \
    "${repo}"
fi
git -C "${repo}" fetch --depth 1 origin "${revision}"
git -C "${repo}" checkout --detach FETCH_HEAD

cd "${repo}"
CLQR_PRECISIONS=FP64 \
CLQR_CUDA_ARCH=60 \
CLQR_BENCHMARK_REPEATS=5 \
CLQR_SANITIZER_TOOLS="memcheck initcheck racecheck synccheck" \
bash scripts/notebook_cuda.sh 2>&1 | \
  tee /kaggle/working/clqr_p100_fp64_report.txt
```

The complete machine description, validation output, and benchmark CSV are saved as `/kaggle/working/clqr_p100_fp64_report.txt`. Replace `revision` with the validated merge commit after the CUDA backend is merged.
